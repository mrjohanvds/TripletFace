- ### Basic Training

The training with no parameters ```!python3 -m tripletface.train -m model -s ../dataset/``` was way too long. About half an hour per epoch so ~5h in total. In order to accelerate to process, I decided to play with parameters.



*   ```-e 5``` to change the number of epoch, diminution of the half of the default value.
*   ```-b 64``` to change the batch size, the number sample processed once. The default value is 32. More the batch size is big, less the noise is. 

*Processing Time with* ```!python3 -m tripletface.train -m model -s ../dataset/ -e 5 -b 64``` *: 2h 08min*

- ### JIT compile the model

JIT means Just In Time. The goal here is to use our previously made model in order to be more efficient

````
%cd TripletFace/

import torch
from tripletface.core.model import Encoder

model = Encoder(64)
weights = torch.load("model/model.pt")['model']
model.load_state_dict( weights )
jit_model = torch.jit.trace( model, torch.rand(3, 3, 3, 3) )
torch.jit.save( jit_model, "model/jitcompile.pt" )
````
